```python
# libraries
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import seaborn as sns
sns.set_style("white")
import pandas as pd
my_dpi=96
```

```python
# Get the data (csv file is hosted on the web)
url = 'https://python-graph-gallery.com/wp-content/uploads/gapminderData.csv'
data = pd.read_csv(url)
#data=pd.read_csv('gapminderData.csv')
data.head()
```

**Definir a variável "url" para importar o ficheiro "gapminderData.csv" (url = 'https://python-graph-gallery.com/wp-content/uploads/gapminderData.csv') no qual se encontram os dados para formular a tabela.**
De seguida usar data = pd.read_csv(url) para ler o ficheiro csv e por fim, utilizar a função data.head() para mostrar, por definição, as primeiras 5 linhas da tabela.


```python
# And I need to transform my categorical column (continent) in a numerical value group1->1, group2->2...
data['continent']=pd.Categorical(data['continent'])
```

Através do **pd.Categorical(data["continent"])** procuramos ler, dentro do ficheiro, a informação **data** da coluna dos continentes.
Ao **igualar data["continent] a pd.Categorical(data["continent"])**, fazemos a **contagem de "continent" diferentes**, **separando** esta informação **em apenas 5 valores**.

```python
anos=data.year.unique()
```

Definir a variável **anos** através da contagem de anos únicos, dentro da informação previamente importada:(**data.year.unique()**).

```python
%matplotlib notebook
%matplotlib notebook

import matplotlib.patches as mpatches

# initialize a figure
#fig=plt.figure(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi)
fig, ax = plt.subplots(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi)

# Add titles (main and on axis)
plt.yscale('log')
plt.xlabel("Life Expectancy")
plt.ylabel("GDP per Capita")
plt.ylim(100,100000)
plt.xlim(30, 90)

def animate(i):
    plt.title("Year: "+str(anos[i]) )
    tmp=data[ data.year == anos[i] ]
    scut=plt.scatter(tmp['lifeExp'], tmp['gdpPercap'] , s=tmp['pop']/200000 , c=tmp['continent'].cat.codes, cmap="Accent", alpha=0.6, edgecolors="white", linewidth=2)
    
ani = animation.FuncAnimation(fig, animate, frames=len(anos), interval=200, blit=False, repeat=False)
    

f = open('gapminderData.csv')
gapminderDataDF = pd.read_csv(f)


plt.plot([1, 2, 3], label="Africa", marker='o', color="w", markerfacecolor='lightgreen', markersize=10)
plt.plot([3, 2, 1], label="Americas", marker='o', color='w', markerfacecolor='moccasin', markersize=10)
plt.plot([3, 2, 1], label="Europe", marker='o', color='w', markerfacecolor='royalblue', markersize=10)
plt.plot([3, 2, 1], label="Oceania", marker='o', color='w', markerfacecolor='whitesmoke', markersize=10)
plt.plot([3, 2, 1], label="Asia", marker='o', color='w', markerfacecolor='chocolate', markersize=10)
plt.legend(bbox_to_anchor=(0.79, 0.25, 0.3, 0.1), loc='upper left', title="Continents")


plt.show()
```

Inicialmente **chamamos a biblioteca de software "matplotlib"** através de "%matplotlib notebook".

Após isto, vamos definir a largura dos eixos em relação ao plot apresentado com 680 dpi de comprimento e 480 dpi de largura através do seguinte código **"fig, ax = plt.subplots(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi)"**.

#### Para **adicionar valores aos eixos X e Y**  e limitá-los utilizamos o seguinte código:

**plt.yscale('log')** --> para apresentar os valores do eixo Y com base em logaritmo e não número integral (por exemplo 10 elevado a 2 em vez de 100)
**plt.xlabel("Life Expectancy")** --> para definir a legenda do eixo x como "Life Expectancy"
**plt.ylabel("GDP per Capita")** --> para definir a legenda do eixo y como "GFP per Capita"
**plt.ylim(100,100000)** --> limitar o eixo Y ao intervalo de valores entre 100 e 100000
**plt.xlim(30, 90)** -->limitar o eixo X ao intervalo de valores entre 30 e 90**

#### Para **definir a animação da apresentação dos dados**, tendo em conta o que foi previamente definido (escalas, valores e informação importada), **utilizamos a função "def animate(i)"**

**def animate(i):** -->iniciar e definir a animaçãoa apresentada
    **plt.title("Year: "+str(anos[i]) )** --> colocar titulo no plot, tendo em conta Year + o valor do ano correspondente em relação ao decorrer da animação
    **tmp=data[ data.year == anos[i] ]** --> colocar determinado tempo para apresentar os anos e o decorrer dos mesmos tendo em base a função
    **scut=plt.scatter(tmp['lifeExp'], tmp['gdpPercap'] , s=tmp['pop']/200000 , c=tmp['continent'].cat.codes, cmap="Accent", alpha=0.6, edgecolors="white", linewidth=2)** --> Através desta função definimos não só os dados a ser importados em relação ao eixo X e Y, para as "bolas", bem como a sua cor e o contorno.

#### Para **definir os frames da animação previamente definida**:

**ani = animation.FuncAnimation(fig, animate, frames=len(anos), interval=200, blit=False, repeat=False)** --> recorremos à função, de maneira a determinar quais serão os frames apresentados durante a animação (frames=len(anos)), apresentando cada frame referente a cada ano, e ainda o intervalo em milisegundos entre os mesmos (interval=200), sem repetições (repeat=False).

**plt.show()** --> apresentar o plot.
