# Questão C2. B: **Temos “bolhas” de cinco (5) cores diferentes. O que representa cada cor? Justifique, com descrição detalhada do processo.**

De acordo com o código apresentado, cada cor representa os seus continentes, estando estes configurados para aparecerem multiplas vezes e o seu tamanho estar dependente da quantidade da população.
Estas bolhas estão ainda programadas para apresentar uma espessura específica através da função "linewidth=2" e uma determinada tonalidade "alpha=0.6".
Na versão "gapminder_animation-FuncAnimation_PLOT_SEM-legend" está presente uma animação contínua, com um conjunto de frames, na qual surgem os frames com um intervalo de tempo de 200 milisegundos, alterando, consequentemente, a frequência com que aparecem estas mesmas bolhas (interval=200).
